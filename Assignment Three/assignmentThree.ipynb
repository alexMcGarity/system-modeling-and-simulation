# Problem 1: 
## Simulate the following continuous-time Lotka-Volterra (predator- prey) model for 0 ≤ t < 50 in Python, with x(0) = y(0) = 0.1, a = b = c = d = 1 and ∆t = 0.01. Visualize the simulation results over time and also in a phase space.
## Then try to reduce the value of ∆t to even smaller values and see how the simulation results are affected by such changes. Discuss what the overall results imply about the Euler forward method.

```python
import matplotlib.pyplot as plt

a = b = c = d = 1.
deltaT = 0.01


def initialize():
    global x, y, xresult, yresult, t, timesteps
    x = 0.1
    y = 0.1
    xresult = []
    yresult = []
    t = 0.
    timesteps = [t]

def observe():
    global x, y, xresult, yresult, t, timesteps
    xresult.append(x)
    yresult.append(y)
    timesteps.append(t)

def update():
    global x, y, xresult, yresult, t, timesteps
    nextx = x + (a * x - b * x * y) * deltaT
    nexty = y + (c * x * y - d * y) * deltaT
    x, y = nextx, nexty
    t += deltaT
```

```python
initialize()
while t < 50:
    update()
    observe()
    
plt.plot(xresult, 'b-', label= 'prey')
plt.plot(yresult, 'g--', label= 'predators')
plt.xlabel('Time steps')
plt.ylabel('Population')
plt.legend()
plt.show()

#lets plot the phase space
plt.plot(xresult, yresult, 'r-')
plt.xlabel('Prey')
plt.ylabel('Predators')
plt.show()
```

# Problem 2: 
## Draw a phase space of the following differential equation (motion of a simple pendulum) in Python:

## $$\frac{d^2\theta}{dt^2} = -\frac{g}{L}\sin\theta$$

```python
#Draw a phase space of the following differential equation (motion of a simple pendulum) in Python:

#d2x/dt2 = -g/l * sin(x)

import matplotlib.pyplot as plt
import math

g = 9.8
l = 100
deltaT = 0.1

def initialize():
    global x, xresult, t, timesteps
    x = 0.1
    xresult = []
    t = 0.
    timesteps = [t]

def observe():
    global x, xresult, t, timesteps
    xresult.append(x)
    timesteps.append(t)

def update():
    global x, xresult, t, timesteps
    nextx = x + (-g/l * math.sin(x)) * deltaT
    x = nextx
    t += deltaT

initialize()
while t < 50:
    update()
    observe()

plt.plot(xresult, 'b-', label= 'x')
plt.xlabel('Time steps')
plt.ylabel('Location')
plt.legend()
plt.show()

#lets plot the phase space
plt.plot(xresult, 'r-')
plt.xlabel('x')
plt.ylabel('dx/dt')
plt.show()
```
