```python
#Conduct a bifurcation analysis of the following dynamical system with parameter r:
#x_t = x_(t-1)^3 -rx_(t-1)

from pylab import *

def initialize():
    global x, xresult
    x = 0.1
    xresult = [x]
    
def observe():
    global x, xresult
    xresult.append(x)

def update():
    global x, xresult
    nextx = x**3 - r*x
    x = nextx

def plot_phase_space():
    initialize()
    for t in range(20):
        update()
        observe()
    plot(xresult)
    axis('image')
    axis([-3, 3, -3, 3])
    title('r = ' + str(r))

rs = [-1.1, -1, 0, 1, 2, 3]
for i in range(len(rs)):
    subplot(1, len(rs), i + 1)
    r = rs[i]
    plot_phase_space()

show()

def plot_cobweb():
    initialize()
    for t in range(20):
        update()
        observe()
    horizontal = [xresult[0]]
    vertical = [xresult[0]]
    for x in xresult[1:]:
        horizontal.append(vertical[-1])
        vertical.append(x)
        horizontal.append(x)
        vertical.append(x)
    plot(horizontal, vertical, 'b')
    axis('image')
    axis([-3, 3, -3, 3])
    title('r = ' + str(r))

rs = [-1.1,-1, 0, 1, 2, 3]
for i in range(len(rs)):
    subplot(1, len(rs), i + 1)
    r = rs[i]
    plot_cobweb()

show()
```

for values of r between -1 and 2, the simulation is stable. For values outside this range the function quickly becomes unpredictable, with r values less than -1.1 exceeding floating point calculations, and r values greater than 3 doing the same. This is because the x**3 portion of the function is not bounded, and scales extremely quickly.

for values of r between -5 and -2, the simulation is in a stable spiral focus. For r values between 0 and 2 the simulation has linear growth, and for r values between 2 and 3 the simulation enters a chaotic spiral.
