Part 2: Implement a simulation code of the following difference equation: 
xt =axt1 +b, x0 =1 

This equation is still linear, but now it has a constant term in addition to axt1. Some real-world examples that can be modeled in this equation include fish population growth with constant removals due to fishing, and growth of credit card balances with constant monthly payments (both with negative b). Change the values of a and b, and see if the systems behaviors are the same as those of the simple exponential growth/decay model. 

For this part, make sure and provide your code and the plots (The visual output of codes regarding to different a and b).

```python
import matplotlib.pyplot as plt

def simulate_difference_equation(a, b, steps):
    x = [1]
    for t in range(1, steps):
        x.append(a * x[t-1] + b)
    return x

def plot_difference_equation(a, b, steps):
    x = simulate_difference_equation(a, b, steps)
    plt.plot(range(steps), x)
    plt.title(f"a={a}, b={b}")
    plt.xlabel("Time (t)")
    plt.ylabel("x(t)")
    plt.show()

plot_difference_equation(1.5, 0, 20)
plot_difference_equation(0.5, 0, 20)
plot_difference_equation(1.5, -0.5, 20)
```

This code defines two functions, simulate_difference_equation and plot_difference_equation. The first function takes as input the values of a, b, and steps, and returns a list x containing the values of x_t for each time step. The second function takes the same inputs and uses the simulate_difference_equation function to generate the values of x_t, and then plots the values using matplotlib.pyplot.

Finally, the code calls plot_difference_equation three times with different values of a and b to show the behavior of the system for different parameters.

From the graphs you can see that the system is very much similar to the simple exponetial growth/decay model. The only difference is that the system has a constant term in addition to axt1.
