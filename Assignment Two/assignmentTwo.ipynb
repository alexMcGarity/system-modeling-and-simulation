```python
import numpy as np
import matplotlib.pyplot as plt

# Define the logistic function
def logistic_growth(N, r, K):
    return r*N*(1 - N/K)

# Define the parameters
r = 2.5
K = 1
N0 = 0.1

# Set up the plot
fig, ax = plt.subplots()
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.plot([0, 1], [0, 1], 'k--', alpha=0.5)  # Plot the y=x line

# Plot the cobweb
N = N0
for i in range(50):
    ax.plot([N, N], [N, logistic_growth(N, r, K)], 'b', alpha=0.5)
    ax.plot([N, logistic_growth(N, r, K)], [logistic_growth(N, r, K), logistic_growth(N, r, K)], 'b', alpha=0.5)
    N = logistic_growth(N, r, K)

# Add labels
ax.set_xlabel('N')
ax.set_ylabel('N(t+1)')
ax.set_title(f'Cobweb Plot of Logistic Growth Model\nr={r}, K={K}, N0={N0}')

# Show the plot
plt.show()
```

```python
from matplotlib.pylab import *
import networkx as nx

g = nx.DiGraph()

for x in range(100):
    g.add_edge((x), (x**x % 100, x))

ccs = [cc for cc in nx.connected_components(g.to_undirected())]
n = len(ccs)
w = ceil(sqrt(n))
h = ceil(n / w)
for i in range(n):
    subplot(int(h), int(w), i + 1)
    nx.draw(nx.subgraph(g, ccs[i]), with_labels = True)

show()
```
